### Outlier Alert

Removing outliers to get my data in range. distributing data into quartiles to find out the IQR in next steps.

### Step 6: Finding IQR

i) Creating two new variables called Q1 and Q3. q1 should contain the 25th percentile for all columns in the dataframe while Q3 should contain the 75th percentile for all the columns in the dataframe.

ii) Calculating the interquartile range **(IQR = Q3 - Q1)** for all columns in the dataframe and printing it to the screen.

```python
Q1= final_data.quantile(0.25)
Q3= final_data.quantile(0.75)
print(final_data.info())

IQR= Q3-Q1
print(Q1)
print(Q3)
print(IQR)
```

---------------------------------------------------------------------------

# Step 7: Identifying Outliers 


i) Calculating how many entries I currently have in the original dataframe.

ii) Using the quartiles and IQR previously calculated, identifying the number of entries I'd have if I had to remove the outliers.

ii) Finding the proportion of outliers that exist in the dataset.


```python

#dataframe = pd.concat(combine)
df = final_data.drop('TIMEFRAME', axis=1)


df_clean = df[~((df < (Q1 - 1.5 * IQR)) |(df> (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_clean.info())


print ("When we have not removed any outliers from the dataset, we have " + str(len(final_data)) + " entries") 


print ("When we have not removed any outliers from the dataset, we have " + str(len(df_clean)) + " entries")
print ("The proportion of outliers which exist when compared to the dataframe are: " + str(len(df_clean)/len(final_data)))
```
